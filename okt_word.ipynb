```python
# 사용자 사전 열기
with open(f"org/openkoreantext/processor/util/noun/names.txt") as f:
    data = f.read()

# 특정 데이터 삭제
# 예시 data.replace('삭제1\n','')
data = data.replace('확진자\n','')

# 파일 새롭게 저장
with open("org/openkoreantext/processor/util/noun/names.txt", 'w') as f:
    f.write(data)
```

```python
# 사용자 사전 열기
with open(f"org/openkoreantext/processor/util/noun/names.txt") as f:
    data = f.read()

# 새로운 단어 추가
# 예시 '추가1\n추가2\n추가3\n'
data += '확진자\n'

# 파일 새롭게 저장
with open("org/openkoreantext/processor/util/noun/names.txt", 'w') as f:
    f.write(data)
```
