```python
# library
import matplotlib.pyplot as plt
```

```python
# 그래프 사이즈 조절
plt.figure(figsize=(8,8))

# 데이터 입력
names= df.columns[2:4]
values = df.iloc[4,[2,3]]

# 파이 차트에 대한 설정 
plt.pie(values, 
        labels=names, 
        labeldistance=1.15, 
        wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' },
        colors= ['tab:blue','tab:orange'],
        autopct = "%0.2f%%",
        textprops = {"fontsize":20},
        startangle = 90,
        explode = [0,0.01],
        shadow = True
       )

# 타이틀 및 축 설정
plt.title('파이 차트', fontsize=25)

# 그래프 보여주기 
plt.show()
```

```python
index_list = []
for i in range(len(df)):
    if df['행정구역별(읍면동)'].loc[i][-3:] == "특별시":
        index_list.append(i)
    elif df['행정구역별(읍면동)'].loc[i][-3:] == "광역시":
        index_list.append(i)
index_list
```

```python
df.iloc[index_list,[0,1]]
```

```python
# 그래프 사이즈 조절
plt.figure(figsize=(18,8))

# 데이터셋 설정
height = df.iloc[index_list,1]
bars = df.iloc[index_list,0]
y_pos = range(len(bars))

# 바 그래프 설정
plt.bar(y_pos, height)

# 축 설정
plt.title('특별시/광역시 인구 바 차트', fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(y_pos, bars, fontsize=15)

# 그래프 보이기
plt.show()
```
