# 경제 일반 빈도 분석

```python
df = pd.read_csv("dataset/naver_news/네이버뉴스_경제_경제일반.csv",index_col=0)
df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
df.drop(df[df['헤드라인'].duplicated(keep='last') == True].index, axis=0, inplace=True)
df.reset_index(inplace=True, drop=True)

noun_list = []
for i in range(len(df)):
    noun_list.append(okt.nouns(df["헤드라인"].loc[i]))
total_word = []

for i in range(len(noun_list)):
    for j in range(len(noun_list[i])):
        total_word.append(noun_list[i][j])
        
stopwords = ['명','세','등','중','제','개','회','것','곳']

result = []
for words in total_word:
    if words not in stopwords:
        result.append(words)
        
counter = collections.Counter(result)
df1 = pd.DataFrame(counter.most_common(30),columns=[['경제','경제'],['단어','빈도수']])
df1
```

## 정치 일반 빈도분석

```python
df = pd.read_csv("dataset/naver_news/네이버뉴스_정치_정치일반.csv",index_col=0)
df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
df.drop(df[df['헤드라인'].duplicated(keep='last') == True].index, axis=0, inplace=True)
df.reset_index(inplace=True, drop=True)

noun_list = []
for i in range(len(df)):
    noun_list.append(okt.nouns(df["헤드라인"].loc[i]))
total_word = []

for i in range(len(noun_list)):
    for j in range(len(noun_list[i])):
        total_word.append(noun_list[i][j])
        
stopwords = ['명','세','등','중','제','개','회','것']

result = []
for words in total_word:
    if words not in stopwords:
        result.append(words)
        
counter = collections.Counter(result)
df2 = pd.DataFrame(counter.most_common(30),columns=[['정치','정치'],['단어','빈도수']])
df2
```

## 특정 단어와 같이 많이 나온 단어 확인하기

### 1. 파일을 불러오고 전처리 진행

```python
df = pd.read_csv("dataset/naver_news/네이버뉴스_사회_사회일반.csv",index_col=0)
df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
df.drop(df[df['헤드라인'].duplicated(keep='last') == True].index, axis=0, inplace=True)
df.reset_index(inplace=True, drop=True)
df
```

### 2. 전처리된 파일 내에서 특정 단어를 포함하는 문장만 추출

```python
temp = []
for i in range(len(df)):
    if '코로나' in df['헤드라인'].loc[i]:
        temp.append(i)
df_temp = df.loc[temp]
df_temp.reset_index(inplace=True, drop=True)
df_temp
```

### 3. 특정 단어를 포함하는 문장들로 빈도분석 진행

```python
df = df_temp
noun_list = []
for i in range(len(df)):
    noun_list.append(okt.nouns(df["헤드라인"].loc[i]))
total_word = []

for i in range(len(noun_list)):
    for j in range(len(noun_list[i])):
        total_word.append(noun_list[i][j])
        
stopwords = ['명','세','등','중','제','개','회','것']

result = []
for words in total_word:
    if words not in stopwords:
        result.append(words)
        
counter = collections.Counter(result)
df4 = pd.DataFrame(counter.most_common(5),columns=[['사회','사회'],['단어','빈도수']])
df4
```
