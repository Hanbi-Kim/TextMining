# 실습: 네이버 영화 랭킹 정보 스크래핑

```python
# 사용할 라이브러리 로드

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# 영화 제목을 포함하는 div태그의 class속성 tit3
soup.find_all("div","tit3")
```

```python
# "a" 태그 안에 'title'속성에 영화 제목이 저장되있음
soup.find_all("div","tit3")[0].find("a")['title']
```

```python
# 타이틀을 담을 title_list 생성
title_list = []
for i in range(len(soup.find_all("div","tit3"))):
    print(soup.find_all("div","tit3")[i].find("a")['title'])
    # 영화 제목 담아주기
    title_list.append(soup.find_all("div","tit3")[i].find("a")['title'])
```

```python
df = pd.DataFrame({"랭킹":range(1,len(title_list)+1),"영화 제목":title_list})
df.set_index("랭킹")
```
