# 노인의 우을증상 시각화 실습 파일

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기 
df = pd.read_csv("dataset/노인의_우울증상.csv", encoding='cp949', header=[1])
df
```

# 1. 2017년 가구 형태별 응답자 수 파이차트 시각화

```python
# 데이터의 콜롬명 확인
df.columns
```

```python
# 특성별(1)이 가구형태별과 같은 데이터만 추출하고, 그 테이블에서 특성별(2)와 응답자수(명) 만 df1에 저장하기
df1 = df[df['특성별(1)']=='가구형태별'][['특성별(2)','응답자수 (명)']]
df1
```

```python
# 응답자수 (명)가 string 타입으로 저장되어 있음으로, int 타입으로 변환
df1['응답자수 (명)'] = df1['응답자수 (명)'].astype(int) 
```

```python
# 응답자수 (명)을 내림차순으로 정렬 
df1.sort_values('응답자수 (명)', ascending=False, inplace=True)
```

```python
df1
```

```python
# 한글 폰트 사용을 위해서 세팅
import matplotlib.font_manager as fm
# font_path = '/usr/share/fonts/AppleGothic.ttf' 
font_path = "C:/Windows/Fonts/malgun.ttf"
font = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font)
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 그래프 사이즈 조절
plt.figure(figsize=(8,8))

# 데이터 입력
values = df1['응답자수 (명)']
names = df1['특성별(2)']

# 파이 차트에 대한 설정 
plt.pie(values, 
        labels=names, 
        labeldistance=1.15, 
        wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' },
        colors= ['tab:blue','tab:orange','tab:green','tab:red'],
        autopct = "%0.2f%%",
        textprops = {"fontsize":20},
        startangle = 90,
#         explode = [0.1,0,0,0],
        shadow = True
       )

# 타이틀 및 축 설정
plt.title('2017년 가구 형태별 응답자 수 파이차트', fontsize=25)

# 그래프 보여주기 
plt.show()
```

# 2017년 교육 형태별 정상/우울증 비교 막대그래프

```python
# 특성별(1)이 교육수준별인 데이터만 추출하기
df2 = df[df['특성별(1)']=='교육수준별'][['특성별(2)','정상 (%)','우울증상 (%)']]
df2
```

```python
# 무학 데이터는 없음으로 제거하기
df2.drop(16, inplace=True)
```

```python
# 데이터가 string 타입으로 저장되어 있음으로, flaot 타입으로 변환
df2['정상 (%)'] = df2['정상 (%)'].astype(float) 
df2['우울증상 (%)'] = df2['우울증상 (%)'].astype(float) 
```

```python
df2
```

```python
# 그래프 사이즈 조절
plt.figure(figsize=(18,8))

# set width of bars
barWidth = 0.35
 
# set heights of bars
bars1 = df2['정상 (%)']
bars2 = df2['우울증상 (%)']
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
# Make the plot
plt.bar(r1, bars1, color='tab:blue', width=barWidth, edgecolor='white', label='var1')
plt.bar(r2, bars2, color='tab:orange', width=barWidth, edgecolor='white', label='var2')
 
# Add xticks on the middle of the group bars
plt.title('2017년 교육 형태별 정상/우울증 비교 막대그래프', fontweight='bold', fontsize=30, pad=20)
plt.xticks([r +0.17 for r in range(len(bars1))], df2['특성별(2)'], fontsize=20)
plt.yticks(fontsize=20)

# Create legend & Show graphic
plt.legend(fontsize=20, labels=['정상','우울증상'])
plt.show()
```
